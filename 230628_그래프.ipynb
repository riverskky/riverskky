## (1) KCD 매출금액과 신한카드 배달앱 지출금액 비교

```python
import pandas as pd
import numpy as np

pd.options.display.float_format = '{:.2f}'.format
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
from datetime import datetime
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

### 데이터 불러오기

```python
kcd = pd.read_excel('Desktop/kcd_증감률.xlsx')
kcd
```

```python
kcd.info()
```

```python
shcard = pd.read_excel('Desktop/shcard_증감률.xlsx')
shcard
```

```python
shcard.info()
```

```python
# 강원도 비교
kcd1 = kcd[ kcd['지역'] == '강원도']
shcard1 = shcard[ shcard['지역'] == '강원도']

fig, ax = plt.subplots(figsize=(15, 6), facecolor='white')

sns.lineplot(x = '날짜', y = '증감률', color = 'blue', ax=ax, data = kcd1)
sns.lineplot(x = '날짜', y = '증감률', color = 'red', ax=ax, data = shcard1)

plt.title('강원도 [KCD 매출 - 신한카드 지출] 증감률 비교', fontsize= 14)
plt.legend(labels=['KCD','신한카드'])

ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))

plt.show()
```

```python
# 상관계수
kcd1 = kcd1['증감률'][:143].reset_index(drop=True)
shcard1 = shcard1['증감률'].reset_index(drop=True)
print(np.corrcoef(kcd1,shcard1)[0,1])
```
