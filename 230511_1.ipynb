```python
a = reg22['사업자등록번호'].dropna().unique()
```

```python
df1 = pd.DataFrame( a, columns = ['number'])
df1
```

```python
noreg22 = noreg22[ (noreg22['등록일자'] >= 202210) ].reset_index(drop=True) # 10월부터 12월까지만 포함
noreg22
```

```python
b = noreg22['사업자등록번호'].dropna().unique()
```

```python
df2 = pd.DataFrame( b, columns = ['number'])
df2
```

```python
def GetDifferences(df1, df2):
    df = pd.concat([df1, df2]).reset_index(drop=True)
    idx = [diff[0] for diff in df.groupby(list(df.columns)).groups.values() if len(diff) == 1]
    return df.reindex(idx)
```

```python
GetDifferences(df1, df2)
```
